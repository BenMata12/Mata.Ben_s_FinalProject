```python
#initially displaying my data
import pandas as pd
filepath = 'Data_Columbia.xlsx'
df = pd.read_excel('Data_Columbia.xlsx')
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#generating a figure showing a comparison of chinook salmon vs. chum vs. Coho spawner numbers in the columbia river vs time.

#extracting data for the species
year = df['YEAR']
chinook_salmon=df[df['SPECIES'] == 'Chinook salmon']
coho_salmon=df[df['SPECIES'] == 'Coho salmon']
chum_salmon=df[df['SPECIES'] == 'Chum salmon']

#defining my x and y axis by grouping the years and the types of salmon together
chinook_yearly = chinook_salmon.groupby('YEAR')['NUMBER_OF_SPAWNERS'].sum()
coho_yearly = coho_salmon.groupby('YEAR')['NUMBER_OF_SPAWNERS'].sum()
chum_yearly = chum_salmon.groupby('YEAR')['NUMBER_OF_SPAWNERS'].sum()

plt.figure(figsize=(50, 25))
plt.plot(chinook_yearly.index, chinook_yearly, label='Chinook')
plt.plot(coho_yearly.index, coho_yearly, label='Coho')
plt.plot(chum_yearly.index, chum_yearly, label='Chum')

plt.xlabel('Year', fontsize=30)
plt.ylabel('Number of Spawners',fontsize=30)
plt.title('Number of Spawners Over Time',fontsize=35)
plt.yticks(range(0, int(chinook_yearly.max()) + 10000, 10000), fontsize=25)#adjusting my tick size so the graph is easier to interpret.
plt.xticks(chum_yearly.index, rotation=45, fontsize=20)
plt.legend(fontsize=35)
plt.show()
```

```python
#next I wanted to find what effect hydroelectric power had on chinook salmon runs specifically.
#Most willamette river runs are generally un-touched, the dams are built upstream of spawning grounds typically
#I wanted to generate two figures each comparing a heavily impeded salmon river run to an unimpeded willamette river run.

popid_104 = df[df['NMFS_POPID'] == 104] #this filters through the dataframe to find the unimpeded Clackimas river run, on the Willamette

popid_90 = df[df['NMFS_POPID'] == 90] #this filters through and finds my impeded run the Secesh river, which is a tributary to the Salmon river.

# Find the common range of years
common_years = set(popid_104['YEAR']).intersection(popid_90['YEAR'])#part of my stack overflow search.

# Finding shared years between the datasets
popid_104_common = popid_104[popid_104['YEAR'].isin(common_years)] #Stack overflow was really helpful for this, I found the ".isin(common_years)" code splice on there, and it was exactly what I needed for this.
popid_90_common = popid_90[popid_90['YEAR'].isin(common_years)]   # I really wanted my data to start at the same time, but I had usually been doing this via hardcoding.

#For this section I looked on stack overflow to find a command that would put my data together by year 
popid_104_by_year = popid_104_common.groupby('YEAR')['NUMBER_OF_SPAWNERS'].sum()
popid_90_by_year = popid_90_common.groupby('YEAR')['NUMBER_OF_SPAWNERS'].sum()

# Plotting my populations against each other.
plt.figure(figsize=(10, 6))
plt.plot(popid_104_by_year.index, popid_104_by_year, label='NMFS_POPID 104(Unimpeded)')
plt.plot(popid_90_by_year.index, popid_90_by_year, label='NMFS_POPID 90 (Impeded)')

plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Spawners', fontsize=14)
plt.title('Comparison of Salmon Return Rates for NMFS_POPID 104 and 90', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.show()

```
